```python
import os
import pandas as pd


def append_csv_files(folder_path):
    dfs = []  # List to store DataFrames

    files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

    for file in files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        dfs.append(df)

    if not dfs:  # Check if the list of DataFrames is empty
        print(f"No CSV files found in {folder_path}. Skipping.")
        return  # Skip the rest of the function if no CSV files are found

    appended_df = pd.concat(dfs, ignore_index=True)

    output_base_path = "/home/miguel/Documents/10_projects/reGrainable/data/csv/grain_properties_grouped/"
    if not os.path.exists(output_base_path):
        os.makedirs(output_base_path)

    folder_name = os.path.basename(folder_path)
    output_file_name = f"{folder_name}.csv"
    output_path = os.path.join(output_base_path, output_file_name)
    appended_df.to_csv(output_path, index=False)

    print(
        f"All CSV files in {folder_name} have been concatenated and saved to {output_path}"
    )


def process_all_folders(main_folder_path):
    for folder in os.listdir(main_folder_path):
        folder_path = os.path.join(main_folder_path, folder)
        if os.path.isdir(folder_path):
            append_csv_files(folder_path)
        else:
            print(f"{folder_path} is not a directory. Skipping.")


main_folder_path = "/home/miguel/Documents/10_projects/draft-01-microstructure-mapping/data/data_csv/grain_properties"

process_all_folders(main_folder_path)
```

```python
import os
import pandas as pd


# Function to concatenate files into a single DataFrame
def concatenate_files(file_list, directory_path):
    dfs = []
    for file_name in file_list:
        file_path = os.path.join(directory_path, file_name)
        dfs.append(pd.read_csv(file_path))
    combined_df = pd.concat(dfs, ignore_index=True)
    return combined_df


# Save the combined DataFrame to a new CSV file
def save_combined_csv(combined_df, directory_path, event_name):
    output_file_name = f"combined_{event_name}.csv"
    output_path = os.path.join(directory_path, output_file_name)
    combined_df.to_csv(output_path, index=False)
    print(f"Combined CSV for {event_name} saved to {output_path}")
    return output_path


# Function to plot from a single CSV file
def plot_from_csv(file_path):
    dataset = pd.read_csv(file_path)
    plot.area_weighted(dataset["equivalent_diameter"], dataset["area"], binsize="rice")
    plt.title(f"Grain size distribution for {os.path.basename(file_path)}")
    plt.show()


# Path to the directory containing your CSV files
directory_path = (
    "/home/miguel/Documents/10_projects/reGrainable/data/csv/grain_properties_grouped/"
)

# Regular expression to match the number after '36'
pattern = re.compile(r"36(\d+)[a-zA-Z]")
all_csv_files = [file for file in os.listdir(directory_path) if file.endswith(".csv")]

# Group the files
grouped_files = {"Interstadial1": [], "Stadial": [], "Interstadial2": []}

for file_name in all_csv_files:
    match = pattern.search(file_name)
    if match:
        number = int(match.group(1))
        if number <= 54:
            grouped_files["Interstadial1"].append(file_name)
        elif 55 <= number <= 64:
            grouped_files["Stadial"].append(file_name)
        elif 65 <= number:
            grouped_files["Interstadial2"].append(file_name)

# Now, process each climatic event
for event_name, files in grouped_files.items():
    if files:
        combined_df = concatenate_files(files, directory_path)
        combined_file_path = save_combined_csv(combined_df, directory_path, event_name)
        plot_from_csv(combined_file_path)
    else:
        print(f"No files found for {event_name}")
```
